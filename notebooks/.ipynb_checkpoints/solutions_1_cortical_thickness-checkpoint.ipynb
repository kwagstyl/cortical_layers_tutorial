## Notebook to interact with and visualise cortical and laminar thickness

```python
import numpy as np
import nibabel as nb
import matplotlib_surface_plotting as msp
import matplotlib.pyplot as plt
import os
%matplotlib inline
```

```python
#load surface data
data_dir='../data'
surf=nb.load(os.path.join(data_dir,'gray_left_327680.surf.gii'))
vertices, faces = surf.darrays[0].data,surf.darrays[1].data
```

```python
#load thickness data. layers 1-6 & total cortical thickness
thickness = np.load(os.path.join(data_dir,'thicknesses_left.npy'))
total_thickness = thickness[-1]
```

```python
#Plot the distribution of thickness values
plt.hist(total_thickness);
plt.xlabel('cortical thickness (mm)')
```

```python
#plot cortical thickness
msp.plot_surf(vertices,faces,total_thickness,vmin=1,vmax=4,
              cmap='turbo',cmap_label='Cortical\nthickness\n(mm)')
```

```python
#load parcellation
parcellation=np.loadtxt(os.path.join(data_dir,'HCP_left.txt'),dtype=int)
labels = np.loadtxt(os.path.join(data_dir,'HCP_labels.txt'),dtype=str)


#find V1
v1_label = parcellation ==np.where(labels=='V1')[0][0]
```

```python
#Plot V1 on to the surface
msp.plot_surf(vertices,faces,v1_label,vmin=0,vmax=1,cmap='viridis',colorbar=False)
```

## Workshop Task 1
Find the thickest ROI

```python
#calculate mean thickness per ROI
#find the ROI with the largest value
```

```python
mean_thicknesses = np.zeros(len(labels))
for k,roi in enumerate(labels):
    roi_label = parcellation ==np.where(labels==roi)[0][0]
    mean_thicknesses[k] = np.mean(total_thickness[roi_label])
    
labels[np.argsort(mean_thicknesses)]
```

## Workshop Task 2
In which ROI is layer IV thickest?

```python
layer_4_thickness= thickness[3]
mean_4_thicknesses = np.zeros(len(labels))
for k,roi in enumerate(labels):
    roi_label = parcellation ==np.where(labels==roi)[0][0]
    mean_4_thicknesses[k] = np.mean(layer_4_thickness[roi_label])
    
labels[np.argmax(mean_4_thicknesses)]
```
